```python
from pyspark.sql.functions import * 
from pyspark.sql.types import * 
```

### **Data Reading**

```python
df = spark.read.format("parquet")\
          .load("abfss://bronze@databricksaccstg.dfs.core.windows.net/products")
```

```python
df.display()
```

```python
df = df.drop("_rescued_data")
```

```python
df.createOrReplaceTempView("products")
```

### **Functions**

```python
%sql
CREATE OR REPLACE FUNCTION databricks_catalog.bronze.discount_func(p_price DOUBLE) 
RETURNS DOUBLE  
LANGUAGE SQL 
RETURN p_price * 0.90
```

```python
%sql
select product_id, price ,databricks_catalog.bronze.discount_func(price) as discounted_price
FROM products
```

```python
df = df.withColumn("discounted_price",expr("databricks_cata.bronze.discount_func(price)"))
df.display()
```

```python
%sql
CREATE OR REPLACE FUNCTION databricks_catalog.bronze.upper_func(p_brand STRING)
RETURNS STRING 
LANGUAGE PYTHON 
AS 
$$ 
    return p_brand.upper()
$$
```

```python
%sql
SELECT product_id, brand, databricks_cata.bronze.upper_func(brand) as brand_upper
FROM products
```

```python
df.write.format("delta")\
            .mode("overwrite")\
            .option("path","abfss://silver@databricksaccstg.dfs.core.windows.net/products")\
            .save()
```

```python
%sql
CREATE TABLE IF NOT EXISTS databricks_catalog.silver.products_silver 
USING DELTA 
LOCATION 'abfss://silver@databricksaccstg.dfs.core.windows.net/products'
```

```python
%sql
CREATE OR REPLACE FUNCTION databricks_cata.bronze.tablefunc(p_year DOUBLE)
RETURNS TABLE(order_id STRING, year DOUBLE)
LANGUAGE SQL
RETURN 
( SELECT order_id, year FROM databricks_cata.gold.factorders
  WHERE year = p_year )
```

```python
%sql
SELECT * FROM databricks_cata.bronze.tablefunc(2024)
```
